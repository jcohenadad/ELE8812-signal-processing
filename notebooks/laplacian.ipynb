# Edge detection using Laplacian method

In this example we detect edges from an image by applying the Laplacian (2nd derivatives) followed by zero-crossing detection.

```
import matplotlib.pyplot  as plt
from skimage import io
import skimage.filters

# Fetch image from the internet
image = io.imread('https://raw.githubusercontent.com/jcohenadad/ELE8812-signal-processing/main/images/brain.png')

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```
# Compute the Laplacian
print(skimage.__version__)
# im_lapl_s3 = skimage.filters. (image, 2, 10)
im_lapl_s30 = skimage.filters.laplace(image, ksize=300)

fig, axes = plt.subplots(1, 2, figsize=(16, 8))
axes[0].imshow(im_lapl_s3, cmap='gray')
axes[1].imshow(im_lapl_s30, cmap='gray')
for ax in axes:
    ax.axis('off')
# fig.colorbar(plot, ax=ax)
```
